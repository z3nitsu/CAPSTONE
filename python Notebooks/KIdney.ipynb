```python
# Import Libraries needed
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import  RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
# importing our dataset
df = pd.read_csv("KIdney2.csv",index_col=0)
```

```python
# reading the file
df.head()
```

```python
# Checking missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.describe()
```

```python
# Correlation
df.corr()
```

```python
# Separting the data(input and Outcome)
X= df[["Al","Su", "Bu","Sc","Wbcc","Htn"]]
Y = df["Class"]
```

```python
# Data Standardization
scaler = StandardScaler()
```

```python
# Fitting the data
scaler.fit(X)
standardized_data = scaler.transform(X)
print(standardized_data)
```

```python
# Storing standardized_data In X variable.
X = standardized_data
Y = df["Class"]
```

```python
#  Train Test Split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 10)
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Train The Model:
classifier = RandomForestClassifier()
```

```python
# Training the support vector machine  classifier
classifier.fit(X_train, Y_train)
```

```python
# accuracy score on the training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print("The accuracy score of the training data:",training_data_accuracy*100)
```

```python
# accuracy score on the test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print("The accuracy score of the test data:",test_data_accuracy*100)
```

```python
# importing Pickle
import pickle
pickle.dump(classifier, open("Kidney.pkl",'wb'))
```
